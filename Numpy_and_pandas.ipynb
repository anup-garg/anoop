```python
# mathemecical Operation
```

```python
np.random.seed(3)
arr = np.random.randint(1,10,(4,4))
print(arr)
print('*'*20)

print(arr+1)
print('*'*20)
print(arr-1)
print('*'*20)
print(arr*2)
print('*'*20)
print(arr/2)
print('*'*20)
print(arr%2)
print('*'*20)
```

```python
# matrix maltiplication
```

```python
np.random.seed(3)
arr1 = np.random.randint(1,10,(4,4))
print(arr1)
arr2 = np.random.randint(1,10,(4,4))
print('*'*20)
print(arr2)
print('*'*20)
print(np.dot(arr1,arr2))
```
